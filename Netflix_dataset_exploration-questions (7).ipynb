## Convert the duration column values to integer format, and override the values of the original columns.

Example -> 90 min(string) -> 90(integer)

```python
data[['duration']].isnull().sum()
```

```python
data = data[~data.duration.isna()]
```

```python
data = data[~data['duration'].str.contains("Season","season")]
```

```python
data['duration'] = data['duration'].apply(lambda x:x.replace("min",""))
```

```python
data
```

```python
data['duration'] = data['duration'].apply(lambda x:int(x))
```
