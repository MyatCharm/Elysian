```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
data = pd.read_csv("datasets/diabetes.csv")
data.head(10)
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.corr(),annot = True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
p = data.hist(figsize = (20,20))
```

```python
zero_columns = [col for col in data.columns if (data[col].min() == 0 and col not in ["Pregnancies", "Outcome"])]
for col in zero_columns:
    data[col] = np.where(data[col]==0, np.nan, data[col])
```

```python
data.isnull().sum()
```

```python
for col in data.columns:
    data.loc[data[col].isnull(), col] = data[col].mean()
```

```python
data.isnull().sum()
```

## Feature Scaling

```python
df = data.iloc[:,:-1]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
d = scaler.fit_transform(df)
print(d)
data.iloc[:,:-1] = d
```

```python
data
```

## Features and Output

```python
X=data.iloc[:,:-1]
X.shape
```

```python
y=data.iloc[:,-1:]
y.shape
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y,test_size=0.3, random_state = 417)
```

```python
print(f'Trainind Data Size : {X_train.shape} \nTesting Data size  : {X_test.shape}')
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
y_pred1 = logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_pred1,y_test)
```

```python
con_matrix1 = confusion_matrix(y_pred1,y_test)
con_matrix1
```

```python
sns.heatmap(con_matrix1,annot=True,cbar=True,cmap = "Greens")
```

```python
print(classification_report(y_pred1,y_test))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=2)
model.fit(X_train,y_train)
```

```python
y_pred2 = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_pred2,y_test)
```

```python
con_matrix2 = confusion_matrix(y_pred2,y_test)
con_matrix2
```

```python
sns.heatmap(con_matrix2,annot=True,cbar=True)
```

```python
print(classification_report(y_pred2,y_test))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, random_state=46)

# fit the model

forest.fit(X_train, y_train)
```

```python
y_pred3 = forest.predict(X_test)

accuracy_score(y_pred3,y_test)
```

```python
rf = RandomForestClassifier(random_state=46, n_jobs=-1)

params = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [50,100,200,150]
}

from sklearn.model_selection import GridSearchCV

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf,
                           param_grid=params,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
rf_best = grid_search.best_estimator_
rf_best
```

```python
forest2 = RandomForestClassifier(max_depth=10, min_samples_leaf=10, n_estimators=500,
                       n_jobs=-1, random_state=46)
forest2.fit(X_train,y_train)
```

```python
y_pred4 = forest2.predict(X_test)
```

```python
accuracy_score(y_pred4,y_test)
```

```python
con_matrix3 = confusion_matrix(y_pred4,y_test)
con_matrix3
```

```python
sns.heatmap(con_matrix3,annot=True,cbar=True, cmap="Blues")
```

```python
print(classification_report(y_pred4,y_test))
```

```python
print ("Accuracy Score For Models")
print(f'Logistic regression : {accuracy_score(y_pred1,y_test)}\nDecision Tree       : {accuracy_score(y_pred2,y_test)}\nRandom Forest       : {accuracy_score(y_pred4,y_test)}')
```
