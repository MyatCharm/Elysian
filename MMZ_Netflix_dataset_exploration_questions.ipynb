## Convert the duration column values to integer format, and override the values of the original columns.

Example -> 90 min(string) -> 90(integer)

```python
df.dropna(subset=['duration'],inplace=True)
#df = df[~df["duration"].str.contains("Season")] ## seasons drop

#df["duration"].apply(lambda x:x.split(" ")[0])
#df["duration"].apply(lambda x:x.replace("min",""))
#df['duration']=df['duration'].astype(int)
```

```python
df[['duration']].isnull().sum()
```

```python
df = df[~df["duration"].str.contains("Season")]
```

```python
df['duration']= df["duration"].apply(lambda x:x.split(" ")[0])
```

```python
df.dropna(subset=["duration"],inplace=True)
df = df[~df["duration"].str.contains("Season")]
df["duration"]=df["duration"].apply(lambda x:x.strip(" min"))
df.head()
```

```python
df["duration"]=df["duration"].astype(int)
df.head()
```
